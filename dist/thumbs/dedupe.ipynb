```python
import os
from collections import defaultdict
from PIL import Image
import imagehash

# ðŸ”§ CHANGE THIS if needed
folder = "."  # or e.g. "/Users/qp252319/Desktop/thumbs"

hashes = defaultdict(list)

files = [
    f for f in os.listdir(folder)
    if f.lower().endswith((".jpg", ".jpeg", ".png"))
]

print(f"Found {len(files)} image files")

for i, filename in enumerate(files, start=1):
    path = os.path.join(folder, filename)
    print(f"[{i}/{len(files)}] Hashing {filename}...")
    try:
        with Image.open(path) as img:
            h = str(imagehash.average_hash(img))
        hashes[h].append(filename)
    except Exception as e:
        print(f"  !! Error on {filename}: {e}")

dupe_groups = [group for group in hashes.values() if len(group) > 1]

print(f"\nFound {len(dupe_groups)} duplicate groups.")
```

```python
for group in dupe_groups:
    print(" -> ".join(group))
import os

deleted = []

for group in dupe_groups:
    keep = group[0]  # keep the first filename in the group
    to_delete = group[1:]
    print(f"Keeping: {keep}")
    for fname in to_delete:
        path = os.path.join(folder, fname)
        if os.path.exists(path):
            os.remove(path)
            deleted.append(fname)
            print(f"  Deleted: {fname}")

print("\nDone.")
print(f"Deleted {len(deleted)} files:")
for f in deleted:
    print("  ", f)
```
